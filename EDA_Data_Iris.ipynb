```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.title('Jumlah Spesies')
sns.countplot(iris['Species'], palette="vlag")
plt.show()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(12,9))
plt.title('Perbandingan Jenis Spesies Berdasarkan Sepal Length dan Width')
sns.scatterplot(x=iris['SepalLengthCm'], y=iris['SepalWidthCm'], hue = iris['Species'])
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(12,9))
plt.title('Perbandingan Jenis Spesies Berdasarkan Sepal Length dan Width')
sns.scatterplot(x=iris['PetalLengthCm'], y=iris['PetalWidthCm'], hue = iris['Species'])
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(iris.iloc[:,1:6].corr(), annot=True)
```

```python
fig, axes = plt.subplots(2,2, figsize=(15,10))
sns.set_theme(style="ticks")
sns.boxplot(x='Species', y='PetalWidthCm', data=iris, orient='v', ax=axes[0,0], palette="vlag")
sns.boxplot(x='Species', y='PetalLengthCm', data=iris, orient='v', ax=axes[0,1], palette="vlag")
sns.boxplot(x='Species', y='SepalWidthCm', data=iris, orient='v', ax=axes[1,0], palette="vlag")
sns.boxplot(x='Species', y='SepalLengthCm', data=iris, orient='v', ax=axes[1,1], palette="vlag")
```

```python
plt.figure(figsize=(15,10))
for i, col in enumerate(iris.columns[1:5]):
    plt.subplot(2,2,i+1)
    plt.hist(iris[col], color='#89B5AF')
    plt.title(col)
```
